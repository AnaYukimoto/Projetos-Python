```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar o conjunto de dados
df = pd.read_csv('/content/ai_job_market_insights.csv')

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Exibir informações gerais sobre o DataFrame
print(df.info())

# Resumo estatístico do DataFrame
print(df.describe())
```

```python
# Remover linhas com valores ausentes
df_clean = df.dropna()

# Converter colunas para tipos de dados apropriados
df_clean['Salary_USD'] = pd.to_numeric(df_clean['Salary_USD'])
```

```python
# Contagem de valores em uma coluna categórica
print(df_clean['Industry'].value_counts())

# Cálculo de médias agrupadas
print(df_clean.groupby('Industry')['Salary_USD'].mean())
```

```python
# Gráfico de barras
df_clean['Industry'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição de Categorias')
plt.xlabel('Industry')
plt.ylabel('Location')
plt.show()
```

```python
# Gráfico de linhas
plt.plot(df_clean['Salary_USD'], df_clean['AI_Adoption_Level'], color='green')
plt.title('Evolução do Valor ao Longo do Tempo')
plt.xlabel('Salary_USD')
plt.ylabel('AI_Adoption_Level')
plt.show()
```

```python
# Gráfico de dispersão
plt.scatter(df_clean['Job_Title'], df_clean['Company_Size'], color='red')
plt.title('Correlação entre Variáveis')
plt.xlabel('Job_Title')
plt.ylabel('Company_Size')
plt.show()
```
