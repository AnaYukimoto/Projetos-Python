```python
import cv2
from google.colab.patches import cv2_imshow

# Carregar um vdeo de arquivo
video_path = '/content/PortfolioCell.mp4'
captura = cv2.VideoCapture(video_path)

# Definir o nmero de frames a serem exibidos
num_frames_to_show = 10
frame_count = 0

while frame_count < num_frames_to_show:
    ret, frame = captura.read()
    if not ret:
        break

    # Exibir o frame original
    cv2_imshow(frame)

    # Converter para tons de cinza
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2_imshow(gray_frame)

    # Aplicar filtro Canny
    edges = cv2.Canny(frame, 100, 200)
    cv2_imshow(edges)

    # Adicione uma pausa para no sobrecarregar a exibio
    cv2.waitKey(100)

    frame_count += 1

# Liberar a captura e fechar as janelas
captura.release()
cv2.destroyAllWindows()
```
